```python
## 깊이너비 우선 탐색 > 타겟 넘버
from itertools import permutations


def solution(numbers, target):
    numbers_m = []
    
    for i in numbers:
        numbers_m.append(i)
        numbers_m.append(0-i)
    
    numbers_m.sort()
    print(numbers_m)
    
    com = list(permutations(numbers_m, len(numbers)))
    com = set(com)
    com = list(com)
        
#     print('--------------------')
#     print(com)    
        
    answer = 0
    for j in com:
        if sum(j) == target:
            new = []
            for k in j:
                new.append(abs(k))
                if sum(new)==sum(numbers):
                    answer += 1
                    print('--------------------')
                    print(sum(j), j)
    print('--------------------')
    return answer
```
