```python
import pandas as pd
import numpy as np
```

```python
X_train = pd.read_csv('../data/X_train.csv', encoding='cp949')
Y_train = pd.read_csv('../data/y_train.csv')
X_test = pd.read_csv('../data/X_test.csv', encoding='cp949')
```

```python
X_train = pd.DataFrame(X_train)
Y_train = pd.DataFrame(Y_train).iloc[:,1]
X_test = pd.DataFrame(X_test)
X_test_copy = X_test
```

```python
X_train.describe()
```

```python
import sklearn.preprocessing
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
result=lb.fit_transform(X_train['주구매상품'])
result=pd.DataFrame(data=result, columns=['주구매상품'])
result
```

```python
result2=lb.fit_transform(X_train['주구매지점'])
result2=pd.DataFrame(data=result2, columns=['주구매지점'])
result2
```

```python
X_train = X_train.drop(['주구매상품'], axis=1)
X_train['주구매상품'] = result
X_train = X_train.drop(['주구매지점'], axis=1)
X_train['주구매지점'] = result2
```

```python
X_test = X_test.drop(['주구매상품'], axis=1)
X_test['주구매상품'] = result
X_test = X_test.drop(['주구매지점'], axis=1)
X_test['주구매지점'] = result2
```

```python
X_train = X_train.fillna({'환불금액': 0})
X_train.isnull().sum()
```

```python
# # feature selection
# x_train = X_train[['총구매액','최대구매액','내점일수','환불금액']]
# x_test = X_test[['총구매액','최대구매액','내점일수','환불금액']]
x_train = X_train
x_test = X_test
```

```python
# 데이터 분할
from sklearn.model_selection import train_test_split
train_feature, test_feature, train_target, test_target = train_test_split(x_train,Y_train,train_size=0.7, random_state=45)
```

```python
# 데이터 정규화
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train_feature = scaler.fit_transform(train_feature)
test_feature = scaler.transform(test_feature)
x_test = scaler.transform(x_test)
```

```python
train_feature
```

```python
# 모델링
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
model = RandomForestClassifier(random_state=44)
parameter = {"max_depth":[3,4,5]}
clf = GridSearchCV(model,param_grid=parameter,cv=5)
clf.fit(train_feature,train_target)
clf_best = clf.best_estimator_
```

```python
# 정확도 예측
from sklearn.metrics import roc_auc_score
pred_1 = clf_best.predict(test_feature)
print(roc_auc_score(test_target,pred_1)) 
# feature_selection을 안했을 경우 : 0.73 / ID를 제거했을 때 : 0.69 / 상관계수 높은것만 : 0.67
```

```python

# 결과값 to_csv로 내보내기
pred = clf_best.predict_proba(x_test)[:,1]
print(pred)
# a = pd.DataFrame({'cust_id' : x_test_copy['ID'], 'Predict': pred}).to_csv('0030000.csv',index=False)
```
